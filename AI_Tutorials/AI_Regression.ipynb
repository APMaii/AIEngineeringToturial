## Regression

Data ro migire (experimental)<br>
shoro mikone random yek khat ro  misaze . shoro mikone hey a va b ro taghir mide. har moghe a va b ro k ytaghir mide . mire faseleye khati k sakhte (y = a*x + b ) ro ba noghat hesabb mikone.  <br>
Mean squared error (MSE) error (faseleye beyne noghate experimental , khati ) **2 + sum ,  mean . 

--> Lin1 (a,b) ---> MSE1 <br>
--> Line2 (a,b)---> mSE2 <br>
...... <br>
--> Line n (a,b)---> mSE n <br>


cost function <br> ma bhafeomn ine ke cost function ro minimize konim --> errore kam she<br>
yani khati ra peyda konim k kmatarin MSE ro dashte bashe -> on khat mishe khate aslie ma --> a , b <br>


Y = a*x + b <br> migim k in khat hamon khatie k nature 


**Linear Regression**


````python
import numpy as np

def linear_regression(x,y):

    candidates=[]

    best_mse = 100000000
    best_a = 0
    best_b = 0

    for a in range(0,101):
        for b in range(0,1001):
            prediction = a*x + b 
            error = y - prediction
            mse = np.mean(error**2)

            candidate_dict= {'mse':mse ,'prediction' : prediction , 'error': error , 'a':a ,'b':b }
            candidates.append(candidate_dict)

            if mse<best_mse:
                best_mse= mse
                best_a = a
                best_b = b

    #best MSE
    #for ...

    
    return candidates, best_mse , best_a, best_b

    

temp = np.array([50,60,70,80,90,100])
estehkam = np.array([520,580,723,813,925,996])



candidates , mse , a , b = linear_regression(temp,estehkam)

print('candidates: ' , candidates[0:10])

print('behtarin MSE:', mse)

print('best A : ',a)
print('best B: ', b)

print(f'Line : y = {a}*x + {b} ')

            
````

````python
all_mse = [c['mse'] for c in candidates]

all_a = [c['a'] for c in candidates]

all_b = [c['b'] for c in candidates]


plt.plot(all_a,all_mse)
plt.title('MSE vs a')
plt.ylabel('MSE')
plt.xlabel('a')
plt.grid()
plt.show()




plt.plot(all_b,all_mse)
plt.title('MSE vs b')
plt.ylabel('MSE')
plt.xlabel('b')
plt.grid()
plt.show()
````

````python
from sklearn.linear_model import SGDRegressor

#schotastic gradient descent regression --> Linear regressor


model = SGDRegressor(loss = 'squared_error',learning_rate='constant',eta0=0.0001 , max_iter=10000,random_state=42)
#Loss function -> yani faselye beyne noghato khat haro --> Squared error
#learning rate --> hamon chzii boid k parametr ro taghir midad , zarb dar moshtagh mishdo --> eta0 --> 0.01 constant sabet
#max iter --> iteration --> 1000 
#random state --> model haee momekne random tosh bashe, random mire a , b avalie ro entekhab mikone --> random-state set mikonan dafe badi ham k run krdn javab yeksan bash
#random_staet --> 42 

temp = np.array([50,60,70,80,90,100])
estehkam = np.array([520,580,723,813,925,996])

#agar x 1 bodi bood , yadet bashe reshape(-1,1) elzame on tabe ast . 

#model = .....
#model.fit(x,y)

model.fit(temp.reshape(-1,1), estehkam) #fit , train --> mire linear regresion  --> random a , b gradient descn hoshmand line haye motefavet , loss mse -->kamtrin

````
